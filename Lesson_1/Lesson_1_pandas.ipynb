```python
import pandas as pd
import numpy as np
```

```python
authors = pd.DataFrame({'author_id': [1, 2, 3], 'author_name': ['Тургенев',
                        'Чехов', 'Островский']})
authors
```

```python
book = pd.DataFrame({'author_id': [1, 1, 1, 2, 2, 3, 3], 'book_title': 
                     ['Отцы и дети', 'Рудин', 'Дворянское гнездо', 
                      'Толстый и тонкий', 'Дама с собачкой', 'Гроза', 
                      'Таланты и поклонники'], 'price': [450, 300, 350, 
                                                         500, 450, 370, 290]})
book
```

```python
author_price = pd.merge(authors, book, on = 'author_id')
author_price
```

```python
top5 = author_price.nlargest(5, 'price')
top5
```

```python
author_stat_min = author_price.groupby('author_name').agg({'price': 'min'})
author_stat_max = author_price.groupby('author_name').agg({'price': 'max'})
author_stat_mean = author_price.groupby('author_name').agg({'price': 'mean'})
author_stat_test = pd.merge(author_stat_min, author_stat_max, on = 'author_name')
author_stat = pd.merge(author_stat_test, author_stat_mean, on = 'author_name')
author_stat = author_stat.rename(columns={'price_x':'min_price', 'price_y':'max_price', 'price':'mean_price'})
author_stat
```

```python
author_price.loc[author_price['author_id'].notnull(), 'cover'] = ['твердая', 'мягкая', 'мягкая', 'твердая', 'твердая', 'мягкая', 'мягкая']
author_price
```

```python
book_info = pd.pivot_table(author_price, 
               index = 'author_name', 
               columns = 'cover', 
               values = 'price',
               aggfunc = np.sum)
```

```python
book_info = book_info.fillna(0)
book_info
```

```python
book_info.to_pickle('book_info.pkl')
```
